```python
import pandas as pd
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("data/high_diamond_ranked_10min.csv").dropna()
data.head()
```

```python
def histogram(valuesArray, title="histo"):
    plt.figure()
    plt.title(title)
    plt.hist(valuesArray)
```

```python
histogram(data["blueWins"],"the histogram of wins and loses")
```

## split the data to test and train before anything

```python
train_data, test_data = train_test_split(data,test_size=.2)
```

## zeroR and histograms
since the dataset is almost balanced with the number os loses is almost equal to the number of wins I think that a zeroR algorithm will have an accuracy of 50% like random classification


```python
histogram(train_data['blueWins'], "train hitogram")
histogram(test_data['blueWins'], "test histogram")
```

## storing training and testing data

```python
import os
os
if not os.path.exists("train"):
    os.mkdir("train")
if not os.path.exists("test"):
    os.mkdir("test")

train_data.to_csv('train/train.csv', index=False)
test_data.to_csv('test/test.csv', index=False)

```
